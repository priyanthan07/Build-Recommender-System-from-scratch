```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv("data/movies.csv")
movies.head()
```

```python
ratings = pd.read_csv("data/ratings.csv")
ratings.head()
```

```python
movies_data = pd.merge(ratings, movies, on="movieId")
movies_data.head()
```

```python
movies_data.groupby("title")["rating"].mean().head()
```

```python
movies_data.groupby("title")["rating"].mean().sort_values(ascending=False).head(25)
```

```python
movies_data.groupby("title")["rating"].count().sort_values(ascending=False).head(25)
```

```python
ratings_mean_count = pd.DataFrame(movies_data.groupby("title")["rating"].mean())
ratings_mean_count
```

```python
ratings_mean_count["rating_counts"] = pd.DataFrame(movies_data.groupby("title")["rating"].count())
ratings_mean_count
```

```python
ratings_mean_count["rating"] = round(ratings_mean_count["rating"], 1)
ratings_mean_count
```

```python
pop_recommendations = ratings_mean_count[(ratings_mean_count["rating"]>4.1) & (ratings_mean_count["rating_counts"]>130)].sort_values(by="rating" ,ascending=False)
```
