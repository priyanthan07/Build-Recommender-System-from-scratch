```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors
import pickle 
```

```python
data = pd.read_csv("data_preprocessed.csv")
```

```python
pivot_table_data = data.pivot_table(columns='User-ID', index='Title', values='Book-Rating')
pivot_table_data
```

```python
pivot_table_data.fillna(0, inplace=True)
```

```python
pivot_table_data
```

```python
data_sparse = csr_matrix(pivot_table_data)
data_sparse
```

```python
model =NearestNeighbors(algorithm='brute')
```

```python
model
```

```python
model.fit(data_sparse)
```

```python
distance, sugg = model.kneighbors(pivot_table_data.iloc[237, :].values.reshape(1, -1),n_neighbors=6)
```

```python
pickle.dump(model, open('artifacts/model.pkl', 'wb'))
pickle.dump(books_name, open('artifacts/books_name.pkl', 'wb'))
pickle.dump(data, open('artifacts/data.pkl', 'wb'))
pickle.dump(pivot_table_data, open('artifacts/pivot_table_data.pkl', 'wb'))
```

```python
def recommeder(books_name):
    book_id = np.where(pivot_table_data.index == books_name)[0][0]
    distance, suggestions = model.kneighbors(pivot_table_data.iloc[book_id,:].values.reshape(1, -1),n_neighbors=6)
    sugg_list = []
    for i in range(len(sugg)):
        sugg_list.append(pivot_table_data.index[suggestions[i]])        
    return sugg_list
                       
```

```python
book_name = "Scandalous Miss Howard (Mira Historical Romance)"
results = recommeder(books_name)
for j in results[0]:
    print(j)
```
